```python
import torch
from transformers import BertTokenizer, BertForMaskedLM

# Load pre-trained BERT model and tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertForMaskedLM.from_pretrained('bert-base-uncased')
model.eval()

# Function to calculate word weightage based on its context in the sentence
def calculate_word_weightage(sentence):
    # Tokenize the sentence into words (subwords for BERT)
    tokens = tokenizer.tokenize(sentence)
    token_ids = tokenizer.convert_tokens_to_ids(tokens)
    
    word_weights = {}
    
    # Loop over each word, mask it, and predict its probability
    for idx, token in enumerate(tokens):
        # Mask the word by replacing it with [MASK]
        masked_tokens = tokens[:idx] + ['[MASK]'] + tokens[idx+1:]
        input_ids = tokenizer.convert_tokens_to_ids(masked_tokens)
        input_tensor = torch.tensor([input_ids])
        
        # Get model predictions for the masked word
        with torch.no_grad():
            outputs = model(input_tensor)
            predictions = outputs.logits
        
        # Get the probability of the original word being predicted correctly
        original_token_id = token_ids[idx]
        predicted_prob = torch.softmax(predictions[0, idx], dim=-1)[original_token_id].item()
        
        # Store the token and its predicted probability (weightage)
        word_weights[token] = predicted_prob

    return word_weights

# Example sentence
sentence = "import torch"
from transformers import BertTokenizer, BertForMaskedLM

# Load pre-trained BERT model and tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertForMaskedLM.from_pretrained('bert-base-uncased')
model.eval()

# Function to calculate word weightage based on its context in the sentence
def calculate_word_weightage(sentence):
    # Tokenize the sentence into words (subwords for BERT)
    tokens = tokenizer.tokenize(sentence)
    token_ids = tokenizer.convert_tokens_to_ids(tokens)
    
    word_weights = {}
    
    # Loop over each word, mask it, and predict its probability
    for idx, token in enumerate(tokens):
        # Mask the word by replacing it with [MASK]
        masked_tokens = tokens[:idx] + ['[MASK]'] + tokens[idx+1:]
        input_ids = tokenizer.convert_tokens_to_ids(masked_tokens)
        input_tensor = torch.tensor([input_ids])
        
        # Get model predictions for the masked word
        with torch.no_grad():
            outputs = model(input_tensor)
            predictions = outputs.logits
        
        # Get the probability of the original word being predicted correctly
        original_token_id = token_ids[idx]
        predicted_prob = torch.softmax(predictions[0, idx], dim=-1)[original_token_id].item()
        
        # Store the token and its predicted probability (weightage)
        word_weights[token] = predicted_prob

    return word_weights

# Example sentence
sentence = "'Weed does the same thing to me. Used to be a stoner now it gives me anxiety everytime i hit it."
word_weights = calculate_word_weightage(sentence)

# Display the word and its weightage (probability that it fits the context)
for word, weight in word_weights.items():
    print(f"{word}: {weight:.4f}")

# Display the word and its weightage (probability that it fits the context)
for word, weight in word_weights.items():
    print(f"{word}: {weight:.4f}")
```
